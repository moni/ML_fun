```python
import utils as ut
```

#### get data

```python
train_url = "https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv"
test_url = "https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv"
```

```python
train_data_path = ut.get_data(train_url)
test_data_path = ut.get_data(test_url)
```

```python
!head {train_data_path}
```

```python
!head {test_data_path}
```

```python
names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
label = 'species'
```

### Stochastic Gradient Descent

```python
sgd = ut.lets_play(train_data_path, batch_size=1, optimizer='sgd', names=names, label=label)
sgd.plot_data_points()
sgd.train()
sgd.plot_metrics()
sgd.predict(test_data_path, batch_size=3)
```

### Mini-Batch Gradient Descent

```python
mbgd = ut.lets_play(train_data_path, batch_size=32, optimizer='sgd', names=names, label=label)
mbgd.plot_data_points()
mbgd.train()
mbgd.plot_metrics()
mbgd.predict(test_data_path, batch_size=3)
```

### Batch Gradient Descent

```python
bgd = ut.lets_play(train_data_path, batch_size=120, optimizer='sgd', names=names, label=label)
bgd.plot_data_points()
bgd.train()
bgd.plot_metrics()
bgd.predict(test_data_path, batch_size=3)
```

### Adam

```python
adam = ut.lets_play(train_data_path, batch_size=120, optimizer='adam', names=names, label=label)
adam.plot_data_points()
adam.train()
adam.plot_metrics()
adam.predict(test_data_path, batch_size=3)
```

### RMSprop

```python
rmsprop = ut.lets_play(train_data_path, batch_size=120, optimizer='rmsprop', names=names, label=label)
rmsprop.plot_data_points()
rmsprop.train()
rmsprop.plot_metrics()
rmsprop.predict(test_data_path, batch_size=3)
```
